## Business Case Overview

You're working as a data scientist for a contracting firm that's rapidly expanding. Now that they have their most valuable employee (you!), they need to leverage data to win more contracts.
```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib
from time import sleep
import numpy as np
import pandas as pd
from random import randint
import os
from selenium import webdriver
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import warnings 
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
chromedriver = "./chromedriver/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
# driver = webdriver.Chrome(chromedriver)
```

```python
# Create a driver called "driver."
driver = webdriver.Chrome(executable_path="./chromedriver/chromedriver")
links = []
df = pd.DataFrame({'links':links})

# Loop through pages to extract links to job descriptions
for pageNum in range(0,100):
    driver.get("https://www.mycareersfuture.sg/search?search=data&sortBy=new_posting_date&page={}".format(pageNum))
    sleep(randint(5,10))
# Grab the page source.
    html = driver.page_source
# Beautiful Soup it!
    html = BeautifulSoup(html, 'lxml')
    url = "https://www.mycareersfuture.sg"
    
# Extract links to job posts
    for link in html.find_all('a', {"class":"bg-white mb3 w-100 dib v-top pa3 no-underline flex-ns flex-wrap JobCard__card___22xP3"}):
        try:
            links.append(url+link.get('href'))  
        except: 
            links.append(np.nan)  

# Create a DF to store the data and export links to CSV file
df = pd.DataFrame({'links':links}, index=None).to_csv("links.csv")    
```

```python
url = pd.read_csv("links.csv")
```

```python
# Drop the unnamed column
url.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
# Create lists for jobs data
company = []
job_title = []
location = []
employment_type = []
seniority = []
job_category = []
salary_range = []
salary_type = []
num_of_applications = []
last_posted_date = []
expiry = []
responsibilities = []
requirements = []
company_info = []

# Loop through pages to extract links to job descriptions
driver = webdriver.Chrome(executable_path="./chromedriver/chromedriver")
for i in range (len(url)):
    try: 
        driver.get(url.links[i])
    # Allow page to load   
        sleep(3)
    # Grab the page source.
        html = driver.page_source
    # Beautiful Soup it!
        html = BeautifulSoup(html, 'lxml')

    # Extract section of HTML that contains job information we are after
        job_info = html.find("div", {"class": "bg-white pa4"})
    # Extract job information 
        try: 
            x1 = job_info.find('p', {"name": "company"}).text  
        except: 
            x1 = np.nan         
        company.append(x1)
        try: 
            x2 = job_info.find('h1', {"id":"job_title"}).text    
        except: 
            x2 = np.nan    
        job_title.append(x2)
        try: 
            x3 = job_info.find('p',{'id':"address"}).text    
        except:
            x3 = np.nan    
        location.append(x3)
        try:
            x4 = job_info.find('p',{'id':"employment_type"}).text    
        except: 
            x4 = np.nan
        employment_type.append(x4)
        try: 
            x5 = job_info.find('p',{'id':"seniority"}).text
        except:
            x5 = np.nan
        seniority.append(x5)
        try: 
            x6 = job_info.find('p',{'id':"job-categories"}).text
        except: 
            x6 = np.nan
        job_category.append(x6)
        try:
            x7 = job_info.find('span',{"class":"salary_range"}).text
        except:
            x7 = np.nan
        salary_range.append(x7)
        try: 
            x8 = job_info.find('span',{"class":"salary_type"}).text
        except:
            x8 = np.nan
        salary_type.append(x8)
        try: 
            x9 = job_info.find('span',{"id":"num_of_applications"}).text
        except:
            x9 = np.nan
        num_of_applications.append(x9)
        try: 
            x10 = job_info.find('span',{"id":"last_posted_date"}).text
        except: 
            x10 = np.nan
        last_posted_date.append(x10)
        try: 
            x11 = job_info.find('span',{"id":"expiry_date"}).text
        except: 
            x11 = np.nan
        expiry.append(x11)
        try:
            x12 = job_info.find('div',{"id":"description-content"}).text
        except:
            x12 = np.nan
        responsibilities.append(x12)
        try: 
            x13 = job_info.find('div',{"id":"requirements-content"}).text
        except:
            x13  = np.nan
        requirements.append(x13)
#   Input NaN values for broken links
    except:
        company.append(np.nan)
        job_title.append(np.nan)
        location.append(np.nan)
        employment_type.append(np.nan)
        seniority.append(np.nan)
        job_category.append(np.nan)
        salary_range.append(np.nan)
        salary_type.append(np.nan)
        num_of_applications.append(np.nan)
        last_posted_date.append(np.nan)
        expiry.append(np.nan)
        responsibilities.append(np.nan)
        requirements.append(np.nan)


    df1 = pd.DataFrame({
                        'company':company, 'job_title': job_title, 'location': location, 
                        'employment_type':employment_type, 'seniority': seniority, "job_category": job_category, 
                       "salary_range":salary_range, "salary_type":salary_type, 
                        "num_of_applications": num_of_applications, "last_posted_date": last_posted_date,
                        "expiry": expiry, "responsibilities":responsibilities, "requirements": requirements, 
                       }, index=None).to_csv("jobs.csv")
```
